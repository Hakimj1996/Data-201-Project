```python
# Importing all the libraries needed.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
# Importing video game developers csv file to notebook.
df = pd.read_csv('https://raw.githubusercontent.com/manugroot/datasets/main/video-games-developers.csv')
```

```python
# Show list of all Data types.
df.dtypes
```

```python
# Locate all developers in the 'United Sates'.
df[df['Country'] == 'United States']
```

```python
# Make a new DataFrame for developers in 'California'
df2= df[df['Administrative division'] == 'California']
df2
```

```python
df2.info
```

```python
# Creating a DataFrame for 'Washington'.
df3= df[df['Administrative division'] == 'Washington']
df3
```

```python
df3.info
```

```python
# Creating a DataFrame for 'Illinois'.
df4= df[df['Administrative division'] == 'Illinois']
df4
```

```python
# "Combo" will be both the created DataFrames combined into a new one.
Combo = pd.concat([df2, df3, df4], ignore_index=True)
```

```python
# Searching for a comparison of the City and Year Established.
Combo.groupby(['Est.'])['City'].count()
```

```python
# Using Seaborn to create visualization of comparison above.
sns.displot(data=df, x="City", hue="Est.", multiple="stack")
```
